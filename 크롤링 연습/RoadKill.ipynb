```python
import pprint
import json



with open('dataset_RoadKill.json', 'r', encoding='UTF8') as f:

    json_data = json.load(f)

pprint.pprint(json.dumps(json_data) )
```

```python
url_list=list(json_data.keys())
```

```python
url_list
```

```python
from tqdm import tqdm
```

```python
from selenium import webdriver
import requests
from bs4 import BeautifulSoup
csv_list = [['date' ,'identifier','label','lat','long','name','numberoflndividual','section']]
for i in tqdm(range(0,100)):
    middle_list = []
    options = webdriver.ChromeOptions()
    options.headless = True
    options.add_argument("windows-size=1920*1080")
    options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36")
    browser = webdriver.Chrome(options=options)
    browser.maximize_window()
    url = url_list[i]
    browser.get(url)
    soup = BeautifulSoup(browser.page_source,'lxml')
    data = soup.find("table",attrs={'class':'description'})
    spool = data.find_all("span",attrs={'class':'literal'})
    for i in spool:
        middle_list.append(i.text.split("\n")[0])
    csv_list.append(middle_list)
    print(middle_list)
    browser.quit()
```

```python

import pandas as pd

=pd.DataFrame(csv_list)




.to_csv('.csv',encoding='utf-8-sig',header=False)
```
