# loc 과 iloc 차이 
+ row를 호출할때는 차이가 없다.
+ column을 호출할때 차이가 발생

```python
df.loc[1:]#row
```

```python
df.iloc[1:]#row
```

```python
df.loc[:2,['Survived','Pclass']] #loc은 column명을 제시
```

```python
df.iloc[:2,1:3] #iloc은 index에 대해
```

```python
df[3:5] #index로 인덱싱 가능/ 하지만 df[3]은 에러가 row 인덱싱
```
