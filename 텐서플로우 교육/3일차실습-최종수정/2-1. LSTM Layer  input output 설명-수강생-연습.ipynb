```python
import tensorflow as tf
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences#길이 맞춰줌 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
#Onehot이랑 Embedding이랑 비슷하다고 보머ㅕㄴ됨ㄴ다. 
```

```python
import numpy as np
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
# 문서의 정의
docs = ['glass of orange juice',
       'bottle of mango juice',
       'glass of mango shake',
       'drink bottle of banana shake',
       'I wnat a glass of cold water',
       'The king and the queen',
       'mand and woman']
```

```python
vocab_size =10000 #그냥 원핫보다 차원의 축소가 이루어지네 #부정확할순 있어도
#오호. . . 
```

### 문제1: 원핫인코딩 연습해보자

```python
# 원핫인코딩해보자
#############################################
encoded_docs = [one_hot(d,vocab_size)for d in docs]
#for i in docs:
#    encoded_docs.append(tf.keras.preprocessing.text.one_hot(i,n=vocab_size))
#############################################
```

```python
print(encoded_docs) #one-hot이랑은 좀 다르게 몇번쨰 단어인지를 호출해줌
```

# 단어 임베딩 (Word Embeddings)

```python
embedding_length = 5 # 임베딩 길이 (단어 하나를 5개의 유닛으로 표현해보자.)
max_doc_len = 10  # 문장 길이
```

### 문제2: 패딩을 시도해보자

```python
# 다음의 pad_sequences를 이용해서 10개의 길이가 되도록 패딩 연습해보자. (디폴트인 pre 대신 post를 사용하라) #뒤에 0을 붙이기 // 앞에 붙이기 
# (https://www.tensorflow.org/api_docs/python/tf/keras/preprocessing/sequence/pad_sequences)
# tf.keras.preprocessing.sequence.pad_sequences(
#     sequences, maxlen=None, dtype='int32', padding='pre',
#     truncating='pre', value=0.0)
```

```python
#############################################

encoded_docs =tf.keras.preprocessing.sequence.pad_sequences(encoded_docs,maxlen=max_doc_len, dtype='int32', padding='post',truncating='post', value=0.0)


#############################################
```

```python
print(encoded_docs)  #(None, 10, 1)
```

### 문제3: 5개의 유닛으로 만드는 임베딩 모델을 구축해보자.

```python
#힌트: Embedding() 찾아보자 (https://www.tensorflow.org/api_docs/python/tf/keras/layers/Embedding)
# tf.keras.layers.Embedding(
#     input_dim, output_dim, embeddings_initializer='uniform',
#     embeddings_regularizer=None, activity_regularizer=None,
#     embeddings_constraint=None, mask_zero=False, input_length=None, **kwargs
# )
# 이중 input_dim, output_dim과 input_length를 정의해보자.
```

```python
#############################################
model = Sequential()
model.add(Embedding(vocab_size,embedding_length,input_length=max_doc_len))
#############################################

model.compile('adam', 'mse')
model.summary()
```

### 문제4: encoded_docs를 (배치크기, 문장길이, 임베딩차원)의 모양으로 만들어라.

```python
# 위에서 만든 모델을 이용해 예측하라.
#############################################

output = model.predict(encoded_docs)
print(output.shape)
print(output)

#############################################
```

### 문제5: LSTM 모델을 구축해보라. 

```python
# 은닉상태의 차원은 64로 지정하라.
#############################################

model = Sequential()
model.add(Embedding(vocab_size,embedding_length,input_length=max_doc_len))
model.add(tf.keras.layers.LSTM(64))

#############################################

model.compile('adam', 'mse')
model.summary()
```

### 문제6: 문제5의 LSTM 모델의 출력을 확인하라.

```python
# LSTM의 출력을 산출하고 이의 차원을 확인하라.
#############################################

output = model.predict(encoded_docs)
print(output.shape)
print(output)

#############################################
```

### 문제7: 감성분석을 할 수 있도록 LSTM을 이용한 로지스틱 모델 예측 모델을 만들어라.

```python
# 힌트: 임베딩 -> LSTM -> Dense, 출력의 갯수가 하나이므로 활성함수를 sigmoid로 사용하라.
```

```python
#############################################


model = Sequential()
model.add(Embedding(vocab_size,embedding_length,input_length=max_doc_len))
model.add(tf.keras.layers.LSTM(64))
model.add(tf.keras.layers.Dense(1,activation='sigmoid'))
#############################################

model.compile('adam', 'binary_cross_entropy') #결과가 하나니까 바이너리이다.
model.summary()
```

### 문제8: 문제 7의 확장된 LSTM 모델의 출력을 확인하라

```python
# 확장된 LSTM의 출력을 산출하고 이의 차원을 확인하라.
```

```python
#############################################


output = model.predict(encoded_docs)
#############################################
```
