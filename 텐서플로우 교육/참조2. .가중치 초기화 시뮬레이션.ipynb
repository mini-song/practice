```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## 7개 layer를 가지며 각 layer에 4096개의 node를 갖는 nerual network 구성
dims = [4096]*7
dims
```

```python
outputs = []
```

```python
## data 16개를 생성
x = np.random.randn(16, dims[0])
```

```python
for Din, Dout in zip(dims[:-1], dims[1:]):
    ## 표준편차 0.1로 초기화
    W = 0.1 * np.random.randn(Din, Dout)
    #W = np.random.randn(Din, Dout) / np.sqrt(Din/2)
    ## tanh activation function을 적용하여 결과를 outputs에 추가함
    x = np.tanh(x.dot(W))
    #x = np.maximum(0, x.dot(W))
    outputs.append(x)
```

```python
## 매 layer의 결과 분포 확인
plt.figure(figsize=(22, 3))
for i in range(len(outputs)):
    plt.subplot(1,6,i+1)
    plt.title("Layer {}\n Mean: {:.2f}\n Std: {:.2f}".format(i+1, np.mean(outputs[i]), np.std(outputs[i])))
    plt.hist(np.reshape(outputs[i], -1), bins=20, range=(-1,1))    
plt.show()
```

```python
## xavier initialization 사용

## outputs 초기화
outputs = []
## data 16개를 생성
x = np.random.randn(16, dims[0])
for Din, Dout in zip(dims[:-1], dims[1:]):
    ## xavier initialization
    #W = 0.02 * np.random.randn(Din, Dout)
    W = np.random.randn(Din, Dout) / np.sqrt(Din)
    ## tanh activation function을 적용하여 결과를 outputs에 추가함
    x = np.tanh(x.dot(W))
    #x = np.maximum(0, x.dot(W))
    outputs.append(x)
## 매 layer의 결과 분포 확인
plt.figure(figsize=(22, 3))
for i in range(len(outputs)):
    plt.subplot(1,6,i+1)
    plt.title("Layer {}\n Mean: {:.2f}\n Std: {:.2f}".format(i+1, np.mean(outputs[i]), np.std(outputs[i])))
    plt.hist(np.reshape(outputs[i], -1), bins=20, range=(-1,1))    
plt.show()
```

```python
## relu & xavier initialization 사용

## outputs 초기화
outputs = []
## data 16개를 생성
x = np.random.randn(16, dims[0])
for Din, Dout in zip(dims[:-1], dims[1:]):
    ## xavier initialization
    #W = 0.02 * np.random.randn(Din, Dout)
    W = np.random.randn(Din, Dout) / np.sqrt(Din)
    ## tanh activation function을 적용하여 결과를 outputs에 추가함
    #x = np.tanh(x.dot(W))
    x = np.maximum(0, x.dot(W))
    outputs.append(x)
## 매 layer의 결과 분포 확인
plt.figure(figsize=(22, 3))
for i in range(len(outputs)):
    plt.subplot(1,6,i+1)
    plt.title("Layer {}\n Mean: {:.2f}\n Std: {:.2f}".format(i+1, np.mean(outputs[i]), np.std(outputs[i])))
    plt.hist(np.reshape(outputs[i], -1), bins=20, range=(-1,1))    
plt.show()
```

```python
## relu & he initialization 사용

## outputs 초기화
outputs = []
## data 16개를 생성
x = np.random.randn(16, dims[0])
for Din, Dout in zip(dims[:-1], dims[1:]):
    ## he initialization
    #W = 0.02 * np.random.randn(Din, Dout)
    W = np.random.randn(Din, Dout) / np.sqrt(Din/2)
    ## tanh activation function을 적용하여 결과를 outputs에 추가함
    #x = np.tanh(x.dot(W))
    x = np.maximum(0, x.dot(W))
    outputs.append(x)
## 매 layer의 결과 분포 확인
plt.figure(figsize=(22, 3))
for i in range(len(outputs)):
    plt.subplot(1,6,i+1)
    plt.title("Layer {}\n Mean: {:.2f}\n Std: {:.2f}".format(i+1, np.mean(outputs[i]), np.std(outputs[i])))
    plt.hist(np.reshape(outputs[i], -1), bins=20, range=(-1,1))    
plt.show()
```
